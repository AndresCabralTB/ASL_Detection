```python
#PREPROCESSING

import os
import cv2
import numpy as np

from brisque import BRISQUE

# Specify the input folder containing the images
input_folder = "/Users/HARRYS SYSTEM PC/Documents/Prueba_vision/B"

# Specify the output folder where processed images will be saved
output_folder_p = "/Users/HARRYS SYSTEM PC/Documents/Prueba_vision_output_handSaved2/B"

os.makedirs(output_folder_p, exist_ok=True)

brisque_list = []

# Iterate through files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".jpg"):  # Process only JPG files
        image_path_p = os.path.join(input_folder, filename)
        image_p = cv2.imread(image_path_p)

        # Convert image to grayscale
        gray_image = cv2.cvtColor(image_p, cv2.COLOR_BGR2GRAY)

        # Apply thresholding to segment the hand
        _, thresholded_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)

        # Find contours in the thresholded image
        contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Create a mask for the hand region
        mask = np.zeros_like(image_p)

        # Draw the contours of the hand on the mask
        cv2.drawContours(mask, contours, -1, (255, 255, 255), thickness=cv2.FILLED)

        # Bitwise AND to extract the hand region
        hand_region = cv2.bitwise_and(image_p, mask)

        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path) 
        

        
        obj = BRISQUE()
        brisque_score = obj.score(image)
        
         #Append the value to an array
        brisque_list.append(brisque_score)

       # Generate the output path for the hand region image
        output_path_p = os.path.join(output_folder_p, f"Hand_{filename}")

        # Save image to new folder
        output_path = os.path.join(output_folder_p, f"{filename}")
        
    
        
        
        
        

        print(f"Image: {filename}" + "    -    " + f"BRISQUE Score: {brisque_score}")
        
        # Save the hand region image
        cv2.imwrite(output_path_p, hand_region)
        print("Image: " + f"{filename}" + "   -   Hand region saved")

print("Hand regions have been extracted and saved")

b_max = max(brisque_list)
print("\nMaximum value: " + str(b_max))
#Print the images from max, min and average

b_min = min(brisque_list)
print("Minimum value: " + str(b_min))

b_average = np.mean(brisque_list)
print("Average value: " + str(b_average))
```
